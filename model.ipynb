```python
import pandas as pd
import numpy as np
import warnings 

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('D:\\nlp models\\fake news\\evaluation.csv', sep=(';'))
df
```

```python
df = df.drop(columns= 'Unnamed: 0')
df
```

```python
# Converting everything into lower case

df['text'] = df['text'].apply(lambda x: x.lower())
df['title'] = df['text'].apply(lambda x: x.lower())
df
```

```python
df.isna().sum()
```

```python
# Removing punctuations
import string

def remove_punc(txt) :
    return txt.translate(str.maketrans('','',string.punctuation))

df['title'] = df['title'].apply(remove_punc)
df['text'] = df['text'].apply(remove_punc)
```

```python
def remove_num(txt):
    new = ''
    for i in txt :
        if not i.isdigit():
            new = new + i
    return new

df['title'] = df['title'].apply(remove_num)
df['text'] = df['text'].apply(remove_num)
```

```python
def remove_emoji(txt):
    new = ''
    for i in txt :
        if i.isascii():
            new = new + i
    return new

df['title'] = df['title'].apply(remove_emoji)
df['text'] = df['text'].apply(remove_emoji)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stopwords = set(stopwords.words('english'))
stopwords
```

```python
def remove(txt) :
    words = txt.split()
    cleaned = []
    for i in words :
        if not i in stopwords :
            cleaned.append(i)
            
    return ' '.join(cleaned)

df['title'] = df['title'].apply(remove)
df['text'] = df['text'].apply(remove)
```

```python
# Feature extraction
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

x_data = df['title'] + df['text']
x = x_data.copy()

x = vectorizer.fit_transform(x)
```
